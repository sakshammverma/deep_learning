```python
from tensorflow import keras
from keras.datasets import imdb
```

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
print(X_train[0])
print(y_train[0])
print(X_test[0])
print(y_test[0])
```

```python
from keras.preprocessing.sequence import pad_sequences
X_train = pad_sequences(X_train, maxlen = 50, padding = 'pre')
X_test = pad_sequences(X_test, maxlen = 50, padding = 'pre')
```

```python
y_train[0]
```

```python
X_train.shape
```

```python
from keras import Sequential
from keras.layers import Dense, Embedding, SimpleRNN
model = Sequential()
model.add(Embedding(10000, output_dim = 2, input_length = 50))
model.add(SimpleRNN(32, activation = 'relu', dropout = 0.2, recurrent_dropout=0.2))
# model.add(SimpleRNN(32, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['acc'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs = 5, validation_data=(X_test, y_test))
```
