```python
!pip install mnist
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten,AveragePooling2D
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
model = Sequential()

model.add(Conv2D(6, kernel_size =(5,5), padding = 'valid', activation = 'tanh', input_shape= (32,32,1)))
model.add(AveragePooling2D(pool_size =(2,2), strides= 2, padding ='valid'))


model.add(Conv2D(16, kernel_size=(5,5), padding ='valid', activation='tanh'))
model.add(AveragePooling2D(pool_size = (2,2), strides= 2, padding ='valid'))

model.add(Flatten())

model.add(Dense(120, activation = 'tanh'))
model.add(Dense(84, activation = 'tanh'))
model.add(Dense(10, activation ='softmax'))
```

```python
model.summary()
```

```python
import numpy as np

# Reshape to (num_samples, 28, 28, 1) to add a channel dimension
X_train_reshaped = X_train.reshape(-1, 28, 28, 1)
X_test_reshaped = X_test.reshape(-1, 28, 28, 1)

# Pad images from 28x28 to 32x32 to match the model's input_shape (32,32,1)
# We add 2 pixels of padding on each side (top, bottom, left, right)
X_train_padded = np.pad(X_train_reshaped, ((0,0),(2,2),(2,2),(0,0)), 'constant', constant_values=0)
X_test_padded = np.pad(X_test_reshaped, ((0,0),(2,2),(2,2),(0,0)), 'constant', constant_values=0)

# Normalize pixel values to range [-1, 1] for tanh activation
X_train_normalized = (X_train_padded / 127.5) - 1
X_test_normalized = (X_test_padded / 127.5) - 1

model.compile(optimizer = 'adam', loss ='sparse_categorical_crossentropy', metrics =['accuracy'])
history = model.fit(X_train_normalized, y_train, epochs = 10, batch_size=64, validation_split = 0.1)
test_loss, test_acc = model.evaluate(X_test_normalized, y_test)
print(f"Test Accuracy: {test_acc:.4f}")
```
