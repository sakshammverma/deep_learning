```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/data_aug.jpg', target_size=(200,200))
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    horizontal_flip=True,
    shear_range=0.2,
    zoom_range=0.2,
    fill_mode='nearest'
)
```

```python
img = image.img_to_array(img)
```

```python
img
```

```python
img.shape
```

```python
input_battch = img.reshape(1, 200, 200, 3) #adding 1 dimension to tell how much image we are gonna transform
```

```python
import os
i= 0
if not os.path.exists('aug'):
    os.makedirs('aug')
for output in datagen.flow(input_battch, batch_size = 1, save_to_dir='aug'):
  i = i + 1
  if i == 10:
    break
```
